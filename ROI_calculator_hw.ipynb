# ROI (Return On Investment) calculator

```python
from IPython.display import clear_output


def welcome():
    print('''
Welcome to ROI Calculator!

This calculator finds your Annual Cash Flow and your Cash-On-Cash Return On Investment (ROI),
based on the values you input for each rental property:

Total Investment
Monthly Income
Monthly Expenses
''')


def calculate():
    operation = input('''Please hit enter to begin entering values.
''')

    # INVESTMENT COSTS
    clear_output()
    down_payment = int(input('\nINVESTMENT--------------------------------------------\n\nPlease enter down payment: '))
    closing_costs = int(input('Please enter closing costs: '))
    rehab_budget = int(input('Please enter rehab budget: '))
    misc_costs = int(input('Please enter any misc. costs: '))
    total_investment = down_payment + closing_costs + rehab_budget + misc_costs
    print(f"\n\tYour TOTAL INVESTMENT is ${total_investment}")
    
    # MONTHLY INCOME
    rental_income = int(input('\n\nMONTHLY INCOME----------------------------------------\n\nPlease enter monthly rental income: '))    
    laundry_income = int(input('Please enter monthly laundry machine income: '))    
    storage_income = int(input('Please enter monthly storage income: '))    
    misc_income = int(input('Please enter any other misc. monthly income: '))    
    total_monthly_income = rental_income + laundry_income + storage_income + misc_income
    print(f"\n\tYour TOTAL MONTHLY INCOME is ${total_monthly_income}")
    
    # MONTHLY EXPENSES
    vacancy = rental_income * 0.05
    property_mgmt = rental_income * 0.1
    print(f"""\n\nMONTHLY EXPENSES--------------------------------------\n
Vacancy budget (5% of monthly rental income): {vacancy}
Property management budget (10% of monthly rental income): {property_mgmt}""")
    tax = int(input('\nPlease enter monthly tax: '))    
    insurance = int(input('Please enter monthly insurance expense: '))    
    utilities = int(input('Please enter monthly utilities expense: '))    
    hoa_fees = int(input('Please enter any monthly HOA fees: '))    
    lawn_snow_fees = int(input('Please enter any monthly budget for lawn care or snow removal: '))    
    repair_budget = int(input('Please enter monthly repair budget: '))    
    capex = int(input('Please enter amount to set aside per month for capital expenditures: '))
    mortgage = int(input('Please enter monthly mortgage payment: ')) 
    total_monthly_expenses = tax + insurance + utilities + hoa_fees + lawn_snow_fees + repair_budget + capex + mortgage + vacancy + property_mgmt
    print(f"""\n\tYour TOTAL MONTHLY EXPENSES are ${total_monthly_expenses}""")
    
    # CASH-ON-CASH ROI
    monthly_cash_flow = total_monthly_income - total_monthly_expenses
    print('\n\nMONTHLY CASH FLOW-------------------------------------\n\nmonthly income - monthly expenses = monthly cash flow')  
    print(f"\n${total_monthly_income} - ${total_monthly_expenses} = ${monthly_cash_flow}")
    print(f"\n\tYour TOTAL MONTHLY CASH FLOW is ${monthly_cash_flow}")
    
    annual_cash_flow = monthly_cash_flow * 12
    print('\n\nANNUAL CASH FLOW--------------------------------------\n\nmonthly cash flow x 12 = annual cash flow')
    print(f"\n${monthly_cash_flow} x 12 = ${annual_cash_flow}")
    print(f"\n\tYour ANNUAL CASH FLOW is ${annual_cash_flow}")

    cash_on_cash_roi = (annual_cash_flow / total_investment) * 100
    print('\n\nCASH-ON-CASH ROI--------------------------------------\n\nANNUAL CASH FLOW / TOTAL INVESTMENT = CASH-ON-CASH ROI')
    print(f"\n${annual_cash_flow} / ${total_investment} = {cash_on_cash_roi}%")
    print(f"\n\tCASH-ON-CASH ROI for this property is {cash_on_cash_roi}%.")

    again()

    
# Define again() function to ask user if they want to use the calculator again

def again():

    calc_again = input('''
Would you like to calculate again?
Please type Y for YES or N for NO.
''')

    if calc_again.upper() == 'Y':
        clear_output()
        calculate()
    elif calc_again.upper() == 'N':
        print('See you later.')
    else:
        again()
        

welcome()
calculate()

#     else:
#         print('You have not typed a valid operator, please run the program again.')
```
